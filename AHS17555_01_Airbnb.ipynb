```python
# task b1,  data loading function 
def task_b1(file_path):
  data_df = pd.read_csv(file_path)
  return data_df
```

```python
# Task b2
def task_b2(data_df):
  print("\n")
  ammeties_list = []
  for i in data_df['amenities']:
    i_list = i.split('"')
    for j in i_list:
      ammeties_list.append(j)

  list1 = [1, 2, 3, 4, 5]
  list2 = ['[', ', ',' ,', ']', '65\\']

  ammeties_list = [x for x in ammeties_list if x not in list2]
  # create a dictionary to count the occurrences of each element
  count_dict = {}
  for element in ammeties_list:
      if element in count_dict:
          count_dict[element] += 1
      else:
          count_dict[element] = 1

  # sort the dictionary by value in descending order
  sorted_dict = sorted(count_dict.items(), key=lambda x: x[1], reverse=True)
  return sorted_dict
```

```python
#Task b3
def task_b3(data_df):
  print("\n")
  location_list =  data_df['host_location'].unique()
  loc_price_avg_list = []
  for i in location_list:
    data_df_location = data_df[data_df['host_location'] == i]
    loc_price_avg = data_df_location['price'].mean()
    loc_price_avg_list.append(loc_price_avg)

  # creating a dataframe using the two lists
  loc_price_avg_df = pd.DataFrame({'Location': location_list, 'Average_price': loc_price_avg_list})
  return loc_price_avg_df
```

```python
#Task b4
def task_b4(data_df):
  print('\n')
  location_list =  data_df['host_location'].unique()

  review_score_ratings_avg_list = []

  for i in location_list:
    data_df_location = data_df[data_df['host_location'] == i]
    review_score_ratings_avg = data_df_location['review_scores_rating'].mean()
    review_score_ratings_avg_list.append(review_score_ratings_avg)

  # create a dataframe using the two lists
  review_score_ratings_avg_df = pd.DataFrame({'Location': location_list, 'Average_review_score_ratings': review_score_ratings_avg_list})
  return review_score_ratings_avg_df
```

```python
#   Task b5 To get the property's location, Name, response rate, instant bookable status and status of superhost with conditions 
#     such as the price should be less than 150 and review score value is greater than 4
def task_b5(data_df):
  print('\n')
  price_b5 = 150
  data_df_price = data_df[data_df['price'] <= price_b5]
  data_df_price_rscore_value = data_df_price[data_df_price['review_scores_value'] >= 4]
  data_b5 = data_df_price_rscore_value[['host_location','name','host_response_rate','instant_bookable','host_is_superhost']]
  return data_b5
```

```python
#Task C1

# create a dictionary to count the occurrences of each element
def task_c1(data_df):
  print('\n')
  #part 1
  count_dict = {}
  for element in data_df['bedrooms']:
      if element in count_dict:
          count_dict[element] += 1
      else:
          count_dict[element] = 1

  # sort the dictionary by value in descending order
  sorted_dict = sorted(count_dict.items(), key=lambda x: x[1], reverse=True)

  # print the top 5 elements with the highest count
  for i in range(len(sorted_dict)):
      print(sorted_dict[i][0], "              :", sorted_dict[i][1])

  # Extract labels and values from the list of tuples
  labels = [x[0] for x in sorted_dict]
  values = [x[1] for x in sorted_dict]

  #Part 2
  # Create a pie chart
  plt.pie(values, labels=labels)

  # Add a title
  plt.title("The proportion of number of Bedrooms")

  # Show the chart
  plt.show()
```

```python
#Task C4
def task_c4(data_df):
  print('\n')
  # extract year from date column
  data_df['year'] = pd.DatetimeIndex(data_df['host_since']).year

  data_df['year'].head()
  data_df_year = data_df[data_df['year'] >= 2019]
  data_df_year = data_df_year[data_df_year['year'] <= 2022]

  data_df_year['year'].unique()

  data_df_year_2022 = data_df_year[data_df_year['year'] == 2022]
  data_df_year_2021 = data_df_year[data_df_year['year'] == 2021]
  data_df_year_2020 = data_df_year[data_df_year['year'] == 2020]
  data_df_year_2019 = data_df_year[data_df_year['year'] == 2019]

  # Create data
  x = np.linspace(0, 10, 100)
  y1 = data_df_year_2022['price']
  y2 = data_df_year_2021['price']
  y3 = data_df_year_2020['price']
  y4 = data_df_year_2019['price']

  # Create subplots for each year
  fig, axs = plt.subplots(2, 2, figsize=(10, 8))

  # Plot Airbnb prices for 2019
  x_2019 = np.linspace(1, 12, 272)
  axs[0, 0].plot(x_2019, y4)
  axs[0, 0].set_title('Airbnb Prices in 2019')
  axs[0, 0].set_xlabel('Month')
  axs[0, 0].set_ylabel('Price')

  # Plot Airbnb prices for 2020
  x_2020 = np.linspace(1, 12, 108)
  axs[0, 1].plot(x_2020,y3)
  axs[0, 1].set_title('Airbnb Prices in 2020')
  axs[0, 1].set_xlabel('Month')
  axs[0, 1].set_ylabel('Price')

  # Plot Airbnb prices for 2021
  x_2021 = np.linspace(1, 12, 138)
  axs[1, 0].plot(x_2021, y2)
  axs[1, 0].set_title('Airbnb Prices in 2021')
  axs[1, 0].set_xlabel('Month')
  axs[1, 0].set_ylabel('Price')

  # Plot Airbnb prices for 2022
  x_2022 = np.linspace(1, 12, 132)
  axs[1, 1].plot(x_2022, y1)
  axs[1, 1].set_title('Airbnb Prices in 2022')
  axs[1, 1].set_xlabel('Month')
  axs[1, 1].set_ylabel('Price')

  # Adjust spacing between subplots
  plt.subplots_adjust(wspace=0.3, hspace=0.5)
  # Display the plot
  plt.show()
```

```python
#c5 To display the number of openings of properties from 2008 to 2022.
def task_c5(data_df):
  print('\n')
  # extract year from date column
  data_df['year'] = pd.DatetimeIndex(data_df['host_since']).year
  # Group the data by a specific column (e.g. year)
  grouped_data_df = data_df.groupby('year')['price'].count()

  grouped_data_df_1 = pd.DataFrame(grouped_data_df, columns=['year','total_openings'])
  grouped_data_df_1['year'] = grouped_data_df.index
  grouped_data_df_1['total_openings'] = grouped_data_df.values
  print(grouped_data_df_1,"\n\n")

  # Create a line plot to display the trend over time
  plt.plot(grouped_data_df_1['year'], grouped_data_df_1['total_openings'])
  plt.title('Total number of openings of properties')
  plt.xlabel('year')
  plt.ylabel('Count')
  plt.show()
```
